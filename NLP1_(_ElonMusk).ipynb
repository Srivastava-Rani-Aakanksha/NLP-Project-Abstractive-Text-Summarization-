```python
import nltk
import re
nltk.download('stopwords')
nltk.download('omw-1.4')
# nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer

ob=PorterStemmer()
lt=WordNetLemmatizer()
```

```python
def lower_case(text):
    text_l=text.lower()
    #print(text)
    return text_l
```

```python
def remove_links(text):
   # Remove all the occurrences of text that ends with .https and .com
  remove_https = re.sub(r'http\S+', '', text)
  remove_com = re.sub(r"\[A-Za-z]*\.com", "", remove_https)
  return remove_https
```

```python
def remove_hashtags(text):
  # Remove all hashtags
  remove_hashtags =re.sub("#[A-Za-z0-9_]+","", text)
  remove_hashtags_c =re.sub("@[A-Za-z0-9_]+","", remove_hashtags )
  return remove_hashtags_c
```

```python
def processing(text):
    #text=regularization(text)
    text=remove_links(text)
    text=lower_case(text)
    text=remove_hashtags(text)
    #text=tokenize(text)
    #print(text)
    # text=removestopword(text)
    # text=stemming(text)
    # text=" ".join(text)
    return text
```
